Covid-19 data analysis & visualization

```python
# -----------------------------
# STEP 1: Import Required Libraries
# -----------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("STEP 1: Libraries Imported Successfully\n")

# -----------------------------
# STEP 2: Dataset Creation
# -----------------------------
np.random.seed(1)
rows = 300

data = {
    "Date": pd.date_range(start="2023-01-01", periods=rows),
    "Country": ["India"] * rows,
    "Confirmed_Cases": np.random.randint(100, 50000, rows),
    "Deaths": np.random.randint(1, 1000, rows),
    "Recovered": np.random.randint(50, 48000, rows),
    "Vaccinated": np.random.randint(1000, 100000, rows)
}

df = pd.DataFrame(data)

print("STEP 2: Dataset Created")
print("Total Records:", df.shape[0])
print("Columns:", list(df.columns))
print("\nSample Data:")
print(df.head(), "\n")

# -----------------------------
# STEP 3: Add Missing Values
# -----------------------------
df.loc[15:20, "Recovered"] = np.nan
df.loc[30:35, "Deaths"] = np.nan

print("STEP 3: Missing Values Introduced (For Cleaning Demo)")
print(df.isnull().sum(), "\n")

# -----------------------------
# STEP 4: Data Cleaning
# -----------------------------
df["Recovered"].fillna(df["Recovered"].mean(), inplace=True)
df["Deaths"].fillna(df["Deaths"].mean(), inplace=True)

print("STEP 4: Data Cleaning Completed")
print("Missing Values After Cleaning:")
print(df.isnull().sum(), "\n")

# -----------------------------
# STEP 5: Feature Engineering
# -----------------------------
df["Active_Cases"] = df["Confirmed_Cases"] - (df["Deaths"] + df["Recovered"])
df["Recovery_Rate (%)"] = (df["Recovered"] / df["Confirmed_Cases"]) * 100
df["Death_Rate (%)"] = (df["Deaths"] / df["Confirmed_Cases"]) * 100

print("STEP 5: New Features Added")
print("New Columns Added:")
print(df[["Active_Cases", "Recovery_Rate (%)", "Death_Rate (%)"]].head(), "\n")

# -----------------------------
# STEP 6: Data Analysis
# -----------------------------
print("STEP 6: Statistical Analysis")
print(df.describe(), "\n")

# -----------------------------
# STEP 7: Data Visualization
# -----------------------------

print("STEP 7: Generating Graphs\n")

plt.figure()
plt.plot(df["Date"], df["Confirmed_Cases"])
plt.title("COVID-19 Confirmed Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Confirmed Cases")
plt.show()

plt.figure()
plt.plot(df["Date"], df["Deaths"])
plt.title("COVID-19 Deaths Over Time")
plt.xlabel("Date")
plt.ylabel("Deaths")
plt.show()

plt.figure()
plt.plot(df["Date"], df["Recovered"])
plt.title("COVID-19 Recovered Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Recovered")
plt.show()

plt.figure()
plt.hist(df["Confirmed_Cases"], bins=30)
plt.title("Distribution of Confirmed Cases")
plt.xlabel("Confirmed Cases")
plt.ylabel("Frequency")
plt.show()

# -----------------------------
# STEP 8: Save Clean Dataset
# -----------------------------
df.to_csv("covid19_cleaned_data.csv", index=False)

print("STEP 8: Cleaned Dataset Saved Successfully")
print("File Name: covid19_cleaned_data.csv\n")


```
